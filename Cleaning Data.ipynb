```python
# Dependencies
import pandas as pd

# Study data files
imdb = "IMDB.csv"

# Read the csv
imdb_data = pd.read_csv(imdb)
```

```python
# Create a data frame out of the csv data
imdb_data_df = pd.DataFrame(imdb_data)
```

```python
# Show the new data frame
imdb_data
```

```python
# Drop unneeded columns
# imdb id, date published, actors, budget, income, metascore, reviews from users/critics
 
imdb_data_df = imdb_data_df.drop(['imdb_title_id', 'original_title', 'date_published', 'budget', 'usa_gross_income', 
                                 'worlwide_gross_income', 'metascore', 'reviews_from_users', 'reviews_from_critics'], axis=1)
```

```python
# Show clean df
imdb_data_df.head()
```

```python
# Drop any 
imdb_data_df = imdb_data_df.dropna()
```

```python
# Show new table
imdb_data_df.head()
```

```python
# Export the CSV
imdb_data_df.to_csv('IMDB_clean.csv')
```
